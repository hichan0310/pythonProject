```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.random.random(100)
b=a*2+(np.random.randn(100)-0.5)*0.1
plt.scatter(a, b)
```

```python
p=0
lr=0.01
```

```python
for i in range(100):
    cost=0
    for j in range(100):
        costt=a[j]*p-b[j]
        cost+=costt
        p -= costt*a[j]*lr
    print(cost/100)
```

```python
plt.scatter(a, b)
plt.plot([0, 1], [0, p])
```
