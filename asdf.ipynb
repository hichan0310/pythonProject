```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 데이터 생성
x = np.random.rand(100)*10
y = 3/(1+math.e**(-2*(x-5)))+np.random.randn(100)*0.1
# y=3/(1+e**(-2x+10))
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
a, b, c=-1, 1, 1
xx=np.array(list(range(1000)))/100
plt.plot(xx, c/(1+math.e**(a*xx+b)))
```

```python
lr=0.001
for i in range(10000):
    cost = 0
    for j in range(100):
        t = c/(1+math.exp(a*x[j]+b))
        cost += (t-y[j]) ** 2
        a -= (t-y[j]) * t * t * (-x[j]*math.exp(a*x[j]+b))/ c * lr
        b -= (t-y[j]) * t * t * (-math.exp(a*x[j]+b)) / c * lr
        c -= (t-y[j]) * t / c * lr
    if i%1000==0:print(cost / 100)
```
