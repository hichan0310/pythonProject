```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 데이터 생성
x = np.random.rand(100) * 10
y = 3 / (1 + math.e ** (-2 * (x - 5))) + 1 + np.random.randn(100) * 0.1
# y=3/(1+e**(-2x+10))
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
a, b, c, d = -2, 10, 3, 3
xx = np.array(list(range(1000))) / 100
plt.plot(xx, c / (1 + math.e ** (a * xx + b)) + d)
```

```python
lr = 0.001
for i in range(10000):
    cost = 0
    for j in range(100):
        t = c / (1 + math.exp(a * x[j] + b)) + d
        cost += (t - y[j]) ** 2
        a -= (t - y[j]) * (t - d) * (t - d) * (-x[j] * math.exp(a * x[j] + b)) / c * lr
        b -= (t - y[j]) * (t - d) * (t - d) * (-math.exp(a * x[j] + b)) / c * lr
        c -= (t - y[j]) * (t - d) / c * lr
        d -= (t - y[j]) * lr
    print(cost / 100)
```
