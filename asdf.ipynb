```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
# 데이터 생성
x = np.random.rand(100) * 10
y = 3 / (1 + math.e ** (-2 * (x - 5))) + 1 + np.random.randn(100) * 0.1
# y=3/(1+e**(-2x+10))
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
a, b, c, d = -2, 10, 3, 3
xx = np.array(list(range(1000))) / 100
plt.plot(xx, c / (1 + math.e ** (a * xx + b)) + d)
```

```python
lr = 0.001
for i in range(10000):
    cost = 0
    for j in range(100):
        t = c / (1 + math.exp(a * x[j] + b)) + d
        cost += (t - y[j]) ** 2
        a -= (t - y[j]) * (t - d) * (t - d) * (-x[j] * math.exp(a * x[j] + b)) / c * lr
        b -= (t - y[j]) * (t - d) * (t - d) * (-math.exp(a * x[j] + b)) / c * lr
        c -= (t - y[j]) * (t - d) / c * lr
        d -= (t - y[j]) * lr
    print(cost / 100)
```

```python
y = np.array(list(map(lambda a: 1 if random.randint(40, 80) < a else 0, x)))
```

```python
plt.scatter(x, y)
```

```python
a, b, c, d = -1, 1, 1, 1    # 초기 a, b, c, d 값에 따라 성공 여부가 달라질 수도 있습니다.
```

```python
lr = 0.001
for i in range(10000):
    cost = 0
    for j in range(100):
        t = c / (1 + math.exp(a * x[j] + b)) + d
        cost += (t - y[j]) ** 2
        a -= (t - y[j]) * (t - d) * (t - d) * (-x[j] * math.exp(a * x[j] + b)) / c * lr
        b -= (t - y[j]) * (t - d) * (t - d) * (-math.exp(a * x[j] + b)) / c * lr
        c -= (t - y[j]) * (t - d) / c * lr
        d -= (t - y[j]) * lr
    print(cost / 100)
print(a, b, c, d)
```
